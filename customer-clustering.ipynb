```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

#Geração do dataset direto no diretório
data = {
    "Age": [19, 21, 20, 23, 31, 22, 35, 40, 18, 29, 50, 45, 60, 55, 38, 25],
    "Annual_Income": [15, 18, 20, 25, 40, 22, 60, 65, 17, 35, 70, 68, 90, 85, 55, 45],
    "Spending_Score": [39, 81, 6, 77, 40, 76, 6, 94, 3, 73, 14, 15, 29, 40, 60, 50]
    }

df = pd.DataFrame(data)
df
```

```python
#Visualização inicial dos dados
sns.pairplot(df)
plt.show()
```

```python
#Padronização dos dados
scaler = StandardScaler() #padronização dos dados
scaled_data = scaler.fit_transform(df) # média 0 e desvio padrão 1
```

```python
#Método do Cotovelo (Elbow Method)

inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Inércia")
plt.grid(True)
plt.show()
```

```python
#Treinar o modelo com k=3 (valor comum)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)

#Atribuir os rótulos dos clusters ao DataFrame original
df["Cluster"] = kmeans.labels_
df
```

```python
#Silhouette Score (qualidade dos clusters)
score = silhouette_score(scaled_data, kmeans.labels_)
score * 100
```

```python
#Visualização dos clusters

plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df["Annual_Income"],
    y=df["Spending_Score"],
    hue=df["Cluster"],
    palette="viridis",
    s=100
)
plt.title("Clusters de Clientes")
plt.xlabel("Renda Anual (k$)")
plt.ylabel("Spending Score")
plt.grid(True)
plt.show()
```
